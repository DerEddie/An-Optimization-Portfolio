# Facility location prolem

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
rnd = np.random
rnd.seed(429)
n=10 #Kunden
m=5 #Standorte

xc = rnd.rand(n)*200
yc = rnd.rand(n)*100

xf = rnd.rand(m)*200
yf = rnd.rand(m)*200

plt.figure(figsize=(5,5))
plt.scatter(xc,yc,c='g')
plt.scatter(xf,yf,c='b')
```

```python
# N = Kunden i
# M = Standorte j

N = [i for i in range(1, n+1)]
M = [i for i in range(1, m+1)]

A = [(i,j) for i in N for j in M]
c = {(i,j):np.hypot(xc[i-1]-xf[j-1],yc[i-1]-yf[j-1]) for (i,j) in A}


f = {i: rnd.randint(1,10) for i in M}
```

```python
# N = Kunden i
# M = Standorte j
from gurobipy import Model, GRB, quicksum
mdl = Model("FLP")

x = mdl.addVars(A, vtype=GRB.BINARY )
y = mdl.addVars(M, vtype=GRB.BINARY)
mdl.modelSense = GRB.MINIMIZE
mdl.setObjective(quicksum(x[i,j]*c[i,j] for i,j in A) + quicksum(y[j]*f[j] for j in M))
mdl.addConstrs(quicksum(x[i, j] for j in M) == 1 for i in N)
mdl.addConstrs( x[i,j] <= y[j] for i,j in A)


mdl.Params.MIPGap = 0.1
mdl.Params.TimeLimit = 20

mdl.optimize()

active_arcs = [a for a in A if x[a].x > 0.9]
        
```

```python

plt.figure(figsize=(10,10))
plt.scatter(xc,yc,c='g')
plt.scatter(xf,yf,c='b')

plt.figure(figsize=(10,10))
plt.scatter(xc,yc,c='g')
plt.scatter(xf,yf,c='b')
for i,j in active_arcs:
    plt.plot([xc[i-1],xf[j-1]],[yc[i-1],yf[j-1]], c="r", linestyle='dotted')
```
